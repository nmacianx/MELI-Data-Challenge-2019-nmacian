```python
df_PR = pd.read_csv('submission_PR.csv')
df_PR.head()
```

```python
#Filter only Spanish results from Spanish model
df_SP=df_SP[df_SP['language']=='spanish']
df_SP.head()
```

```python
#Filter only Portuguese results from Portuguese model
df_PR=df_PR[df_PR['language']=='portuguese']
df_PR.head()
```

```python
#Concatenate Portuguese and Spanish dataframes
frames = [df_PR, df_SP]
df = pd.concat(frames)
df.head()
```

```python
#Get the right order for the dataframe
df.sort_values(by=['id'],ascending=True,inplace=True)
df.iloc[:100,:]
```

```python
#Cell to check if there are any empty titles
df[df.title.isna()]
```

```python
#Style the submission dataframe as it needs to be for correct submission
df.pop('language');
df.pop('title');
df.head()
```

```python
#Generate submission CSV file
df.to_csv('submission.csv', index=False)
```
