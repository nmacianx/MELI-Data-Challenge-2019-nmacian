```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
#Download training dataset and filter Portuguese titles
!wget https://meli-data-challenge.s3.amazonaws.com/train.csv.gz
!gzip -d train.csv.gz
df = pd.read_csv('train.csv')
df = df[df['language']=='portuguese']
```

```python
#Split dataset into realiable and unreliable according to label
df_r = df[df['label_quality']=='reliable']
df_u = df[df['label_quality']=='unreliable']
```

```python
efim = df.groupby(['category']).count()
```

```python
def subsample_set(df_r, df_u):
    #get category names
    set_r = set(df_r.category.unique())
    set_u = set(df_u.category.unique())
    #get common category names, categories only in reliable and only in unreliable
    cat_ambos = list(set_r & set_u)
    cat_r = list(set_r - set_u)
    cat_u = list(set_u - set_r)
    df_sub = df_r.iloc[0:0]   #get empty dataframe
    #get quantities
    cant_r = df_r.groupby(['category']).count()
    cant_u = df_u.groupby(['category']).count()
    total = len(cat_ambos)+len(cat_r)+len(cat_u)
    i=0
    for cat in cat_ambos:        
        i+=1
        print('It. {} de {}'.format(i,total))
        #get reliable and unreliable quantity for the category
        nro_r = cant_r[cant_r.index==cat]['title'][0]
        nro_u = cant_u[cant_u.index==cat]['title'][0]
        #filter the titles for the category
        rel = df_r[df_r['category']==cat]
        unrel = df_u[df_u['category']==cat]
        #subsample category
        if nro_r>=3800:
            aux, _ = train_test_split(rel, train_size=3800, shuffle=True)
            df_sub = pd.concat([df_sub, aux])
        else:
            df_sub = pd.concat([df_sub, rel])
            faltan = 3800 - nro_r
            if faltan<nro_u:
                aux, _ = train_test_split(unrel, train_size=faltan, shuffle=True)
                df_sub = pd.concat([df_sub, aux])
            else:
                df_sub = pd.concat([df_sub, unrel])
    
    for cat in cat_r:
        i+=1
        print('\r')
        print('It. {} de {}'.format(i,total))
        nro_r = cant_r[cant_r.index==cat]['title'][0]
        rel = df_r[df_r['category']==cat]
        if nro_r>=3800:
            aux, _ = train_test_split(rel, train_size=3800, shuffle=True)
            df_sub = pd.concat([df_sub, aux])
        else:
            df_sub = pd.concat([df_sub, rel])
                
    for cat in cat_u:
        i+=1
        print('It. {} de {}'.format(i,total))
        nro_u = cant_u[cant_u.index==cat]['title'][0]
        unrel = df_u[df_u['category']==cat]
        if nro_u>=3800:
            aux, _ = train_test_split(unrel, train_size=3800, shuffle=True)
            df_sub = pd.concat([df_sub, aux])
        else:
            df_sub = pd.concat([df_sub, unrel])
    return df_sub
```

```python
df_pr = subsample_set(df_r, df_u)
```

```python
import unicodedata
import pandas as pd
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
import re
from bs4 import BeautifulSoup
```

```python
def normalize_title(text):
    text = unicodedata.normalize('NFKD', text.lower()).encode('ASCII', 'ignore').decode('utf8')
    
    REPLACE_BY_SPACE_RE = re.compile('[/(){}\[\]\|@,-;._]')
    BAD_SYMBOLS_RE = re.compile('[^0-9a-z ]')
    STOPWORDS = set(stopwords.words('spanish'))
    STOPWORDS1 = set(stopwords.words('portuguese'))
    borrar = pd.Series(["ante","kit","cuotas","cuota","ctas","bajo","cabe","kg","con","contra","de","desde","durante","en","entre","hacia","hasta","mediante","para","por","segun","sin","so","sobre","tras","versus","ante","apos","ate","com","contra","de","desde","em","entre","para","por","perante","sem","sob","sobre","tras","marron","gris","rosa","naranja","morado","negro","celeste","blanco","azul","verde","rojo","amarillo","vermelho","verde","azul","amarelo","laranja","roxo","branco","cinza","preto","marrom","rosa","lila","original","cm","cms","mini","unidades","gratis","envio","ar","mm","mms","litros","sin","combo","pack","oferta","promocao","profissional","novo","grande","tipo","ml","lts","lt","alta","alto","metros","mts","doble","simple","triple","premium","universal","new","box","marca","micro","power","importado","usado","como","in","excelenta","estado","nova","master","oficial","na","nino","ninos","varios","usa","anti","mas","los","envios","usada","regalo","regalos","gb","largo","angosto","chico","izquierdo","derecho","unidade","gr","pequeno","comoda","precio","unico","unica","excelente","hermoso","bello","impecable","anos"]).tolist()
    borrar_pal = set(borrar)
    
    text = REPLACE_BY_SPACE_RE.sub(' ', text) # replace REPLACE_BY_SPACE_RE symbols by space in text
    text = BAD_SYMBOLS_RE.sub('', text) # delete symbols which are in BAD_SYMBOLS_RE from text
    text = ' '.join(word for word in text.split() if word not in STOPWORDS) # delete stopwords from text
    text = ' '.join(word for word in text.split() if word not in STOPWORDS1) # delete stopwords from text
    text = ' '.join(word for word in text.split() if word not in borrar_pal) # delete stopwords from text
    text = re.sub(r"\s*[^A-Za-z]+\s*", " ", text) #delete symbols
    text = re.sub(r'\b\w{1,1}\b', '', text) #delete one char words
    text = re.sub(r"http\S+", "", text) #delete urls
    text = re.sub("\s\s+", " ", text) #delete extra whitespaces
    return text
```

```python
#clean titles with normalize_title function
df_pr['title'] = df_pr.title.apply(normalize_title)
```

```python
#get rid of empty titles
df_pr = df_pr[~df_pr.title.isna() & (df_pr.title != 'nan') & (df_pr.title != '')]
#Save cleaned and subsampled Portuguese dataset
df_pr.to_csv('train_pr2.csv', index=False)
```
